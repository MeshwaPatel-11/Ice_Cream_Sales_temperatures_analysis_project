```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('Ice Cream Sales - temperatures.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
print("Duplicate rows:")
print(df.duplicated())

df = df.drop_duplicates()
```

```python
df.describe()
```

```python
# Create a histogram of the Ice Cream Profits column
plt.hist(df['Temperature'], bins=20, edgecolor='black')

# Set the title and labels for the plot
plt.title('Ice Cream Profits vs. Temperature')
plt.xlabel('Temperature')
plt.ylabel('Ice Cream Profits')

# Display the plot
plt.show()
```

```python

# Create a histogram with a seaborn
sns.histplot(data=df, x='Temperature', bins=20, edgecolor='black')


plt.title('Ice Cream Profits vs. Temperature')
plt.xlabel('Temperature')
plt.ylabel('Ice Cream Profits')
plt.gca().set_facecolor('lightblue')

plt.show()
```

```python
fig = px.histogram(df, x='Temperature', nbins=20)

# Show the plot
fig.show()
```
